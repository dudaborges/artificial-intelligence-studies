```python
# apaga as colunas desnecessárias. axis especifica que o drop deve ser realizado ao longo das colunas
df = df.drop(['Title', 'Positive Feedback Count', 'Unnamed: 0', ], axis=1)
#apaga as linhas que contém o valor NaN. inplace indica que as alterações serão feitas diretamente no df, ao invés de criar uma cópia
df.dropna(inplace=True)
```

```python
#cria uma coluna nova, com os seguintes valores: positive/neutral/negative de acordo com os valores do rating
df['Polarity_Rating'] = df['Rating'].apply(lambda x: 'Positive' if x > 3 else('Neutral' if x == 3  else 'Negative'))
```

```python
#seleciona todos os valores positivos, mas limita em até 8000 linhas. 
df_Positive = df[df['Polarity_Rating'] == 'Positive'][0:8000]
#seleciona todos os valores neutros e adiociona em um novo dataframe
df_Neutral = df[df['Polarity_Rating'] == 'Neutral']
#seleciona todos os valores negativos e adiociona em um novo dataframe
df_Negative = df[df['Polarity_Rating'] == 'Negative']
```

```python
df_Positive
```

```python
#cria um df balanceado, onde cada classe tenha o mesmo número de dados. replace seleciona mais de uma vez um dado
df_Neutral_over = df_Neutral.sample(8000, replace=True)
df_Negative_over = df_Negative.sample(8000, replace=True)
df = pd.concat([df_Positive, df_Neutral_over, df_Negative_over], axis=0)
```

```python
df
```

```python
def get_text_processing(text):
    #obtem uma lista de palavras de parada em inglês, como "a", "an", "the", que serão removidas no pré-processamento
    stpword = stopwords.words('english')
    #remove todos os caracteres de pontuação
    no_punctuation = [char for char in text if char not in string.punctuation]
    #reune os caracteres sem pountação em uma única string
    no_punctuation = ''.join(no_punctuation)
    #verifica cada palavra se são palavras de paradas
    return ' '.join([word for word in no_punctuation.split() if word.lower() not in stpword])
```

```python
df['review'] = df['Review Text'].apply(get_text_processing)
df.head()
```

```python
df = df[['review', 'Polarity_Rating']]
df.head()
```

```python
#função onde cada categoria forma uma coluna com valores binários
one_hot = pd.get_dummies(df["Polarity_Rating"])
#apaga a coluna antiga
df.drop(["Polarity_Rating"], axis=1, inplace=True)
#concatena o one_hot nas colunas do dataframe principal
df = pd.concat([df, one_hot], axis=1)
df.head()
```

```python
X = df["review"].values
y = df.drop("review", axis=1).values
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42
)
```
